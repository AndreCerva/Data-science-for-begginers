##### Filtrado de datos: obtener información relevante en filas y/o columnas de nuestras bases de datos

```python
import pandas as pd
df= pd.read_csv('datos2008.csv',nrows=10000)
df.columns
```

##### Para obtener información por filas usar la sigueinte sintaxis

```python
df['ArrDelay'].head() #Recordar que head devuelve los 5 primeros valores
```

##### Para obtener información por columnas usar la siguiente sintaxis

```python
df[0:10]
```

#### Filtrar por un valor concreto
escoger df entero y luego filtro

```python
df[df['ArrDelay']>200].head()
```

```python
df[df['ArrTime']>500].head()
```

##### Añadir más de 1 filtro en la misma instruccion

```python
df[(df["ArrTime"]>60)&(df["ArrDelay"]>200)].head() #& para ambas concidiones, | para una o la otra
```

##### Filtrar una ciudad u otra

```python
df[df.Origin.isin(["HOU","SFO"])].head()
```

#### Filtrar datos perdidos

```python
df[pd.isna(df["ArrDelay"])].head()
```

```python
len(df[pd.isna(df["ArrDelay"])])
```
